```python
def combine_dataframes(dframes):
    df = pd.concat(dframes,ignore_index=True)
    df = df.reset_index(drop=True).iloc[:,1:]
    df.drop_duplicates(subset=['key','expire_time_gmt','valid_time_gmt','temp','dewPt'])
    
    return df
```

```python
import os
import pandas as pd
# valids = [x for x in os.listdir('weather_data/') if '.csv' in x]

valids = [x for x in os.listdir() if '.csv' in x]
```

```python
# # 
# os.chdir('../weather_data/')
# # print(os.getcwd())
master_df = combine_dataframes([pd.read_csv(x) for x in valids])
```

```python
master_df = master_df.dropna(subset=['temp'])
master_df.to_csv('master.csv')
```
